## Graficando familia de soluciones

La ecuación diferencial:
$$
x^{\prime \prime}(t)=0
$$

significa que la segunda derivada es cero → la función es lineal.

Al integrar dos veces:

$$
\begin{gathered}
x^{\prime}(t)=A \\
x(t)=A t+B
\end{gathered}
$$


Es decir:
✓ todas las soluciones son rectas

✓ pendiente $=A$

✓ intersección $=B$

cada par $(A,B)$ produce una recta distinta.

```python
import numpy as np
import matplotlib.pyplot as plt

# Dominio de tiempo
t = np.linspace(-5, 5, 400)

# Familia de soluciones: x(t) = A t + B
A_values = [-2, -1, -0.5, 0.5, 1, 2]
B_values = [-2, 0, 2]

plt.figure()
for A in A_values:
    for B in B_values:
        x = A * t + B   # ← aquí sí se incluye B
        plt.plot(t, x)

plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Familia de soluciones de x''(t) = 0")
plt.show()
```
