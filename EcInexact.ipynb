## Planeen y resuelvan:

En una comunidad rural se estudia un indice de riesgo ajustado al sarampion $S(x, y)$, donde una variable representa el nivel acumulado de una campaña preventiva y la otrarepresenta el nivel de exposición al riesgo.

Se sabe que un pequeño cambio en el índice de riesgo depende de los cambios entre variables según:
- el cambio marginal respecto a una es proporcional a $y(1+x)$,
- el cambio marginal respecto a la otra es proporcional a $x$.

Además, se desea describir las trayectorias ( $x, y$ ) para las cuales el riesgo ajustado permanece constante.



a) 
1. Plantea la ecuación diferencial que modela las trayectorias de riesgo constante.
2. Verifica si la ecuación es exacta.
3. Si no lo es, encontrar un factor integrante adecuado.
4. Resuelve la ecuación diferencial.

b ) 
Supongamos que al inicio de la observación:
- La intensidad acumulada de la campaña es $x=1$,
- El nivel de exposición medido es $y=2$.

Establece la condicion inicial y obten la solución particular

```python
import numpy as np
import matplotlib.pyplot as plt

# Familia de soluciones: x*y*e^x = C  ->  y = C / (x e^x)
def y_sol(x, C):
    return C / (x * np.exp(x))

# Dominio: evitamos x = 0
x = np.linspace(0.2, 3.0, 600)

# Valores de C para la familia
C_values = [0.5, 1, 2, 2*np.e, 8, 12]

plt.figure(figsize=(9,5))

# Familia de soluciones
for C in C_values:
    plt.plot(x, y_sol(x, C), color="0.75", linewidth=1)

# Solución particular: y(1)=2  -> C = 2e
C_part = 2 * np.e
y_part = y_sol(x, C_part)

plt.plot(x, y_part, color="blue", linewidth=3,
         label=r"Solución particular: $y(1)=2$")

# Punto inicial
plt.scatter([1], [2], color="red", s=60, zorder=5, label="Punto inicial (1,2)")

plt.xlabel("x")
plt.ylabel("y(x)")
plt.title(r"Familia de soluciones de $y(1+x)\,dx + x\,dy = 0$")
plt.grid(True)
plt.legend()
plt.show()
```

Cada curva representa una combinación de campaña preventiva $x$ y nivel de exposición $y$ que mantiene constante el mismo índice de riesgo ajustado.
