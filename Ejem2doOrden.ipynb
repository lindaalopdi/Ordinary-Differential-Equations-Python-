```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 2, 400)


def y_sol(t, C1, C2):
    return C1*np.exp(t) + C2*np.exp(-t)

plt.figure(figsize=(9,5))

#Fam de soluciones
C1_vals = [-2, -1, -0.5, 0.5, 1, 2]
C2_vals = [-1, 0, 1]

for C1 in C1_vals:
    for C2 in C2_vals:
        plt.plot(t, y_sol(t, C1, C2), color="0.85", linewidth=1.1)

# solución particular: y(0)=1, y'(0)=0
y_part = y_sol(t, 0.5, 0.5)

plt.plot(t, y_part, linewidth=3, label="Solución particular", color= "orchid")

#Punto inicial
plt.scatter([0], [1], s=60, zorder=5, c="purple")

plt.xlabel("t")
plt.ylabel("y(t)")
plt.title(r"Familia de soluciones de $y''-y=0$")
plt.grid(True)
plt.legend()
plt.show()
```
