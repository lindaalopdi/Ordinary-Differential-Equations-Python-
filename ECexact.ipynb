## Ecuaciones exactas

### Ejemplo 1:

Tenemos:
$$
(2 x y) d x+\left(x^2\right) d y=0
$$


Sabemos que la función potencial es

$$
F(x, y)=x^2 y
$$

y por tanto la familia de soluciones es

$$
x^2 y=C .
$$


Si tomamos una condición inicial, por ejemplo

$$
(1,2),
$$

entonces

$$
C=1^2 \cdot 2=2
$$

y la solución particular es

$$
y=\frac{2}{x^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Función 
def F(x, y):
    return x**2 * y

x = np.linspace(-5, 5, 800)
y = np.linspace(-5, 5, 800)
X, Y = np.meshgrid(x, y)

Z = F(X, Y)

plt.figure(figsize=(8,6))

# Familia de soluciones ---
C_values = [-4,-3, -2, -1, 1, 2, 4]
cs = plt.contour(X, Y, Z, levels=C_values, linewidths=1)
plt.clabel(cs, inline=True, fontsize=9)

# --- Solución particular (pasa por (1,2)) ---
C_part = 2
cs_part = plt.contour(X, Y, Z, levels=[C_part], linewidths=3)

# Punto inicial
plt.scatter([1], [2], s=80, label="(1,2)")

plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Familia de soluciones de $(2xy)dx+(x^2)dy=0$  ($x^2y=C$)")
plt.grid(True)
plt.legend()
plt.show()
```

### Ejemplo 2:

```python
import numpy as np
import matplotlib.pyplot as plt

# Función de costo
def C(x, y):
    return x**3 + x*y + y**2

# Malla para graficar
x = np.linspace(-2.0, 2.0, 600)
y = np.linspace(-2.0, 2.0, 600)
X, Y = np.meshgrid(x, y)

Z = C(X, Y)

plt.figure(figsize=(8,6))

# Familia de soluciones: niveles K
K_values = [-2, -1, 0, 1, 2, 3, 4, 6]
cs = plt.contour(X, Y, Z, levels=K_values, linewidths=1)
plt.clabel(cs, inline=True, fontsize=9)

#Solución particular: pasa por (1,1)
K_part = C(1, 1)  # = 3
cs_part = plt.contour(X, Y, Z, levels=[K_part], linewidths=3)
# Etiqueta manual para distinguir
plt.clabel(cs_part, inline=True, fontsize=10)

# Punto (1,1)
plt.scatter([1], [1], s=60, label="Punto (1,1)")

plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Familia de soluciones: $C(x,y)=K$ con $C(x,y)=x^3+xy+y^2$ (particular $K=3$)")
plt.grid(True)
plt.legend()
plt.show()
```

En ecuaciones exactas de la forma $M d x+N d y=0$

la solución es $F(x, y)=K$. Por eso aquí la "familia" no es $y(t)$, sino curvas en el plano $(x, y)$.
